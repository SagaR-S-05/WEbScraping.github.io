```python
import requests 
from bs4 import BeautifulSoup as bs 
import pandas as pd 
```

```python
url="https://www.espncricinfo.com/cricketers/virat-kohli-253802" 
print(url) 
r=requests.get(url) 
print(r)
print(r.text) 
```

```python
soup=bs(r.text,"lxml") 
print(soup)
```

```python
table=soup.find("table",class_="ds-w-full ds-table ds-table-md ds-table-bordered ds-border-collapse ds-border ds-border-line ds-table-auto ds-overflow-scroll")
print(table)
title=table.find_all("th") 
print(title)
```

```python
header=[] 
for i in title:
    name=i.text
    header.append(name) 
print(header)
```

```python
df=pd.DataFrame(columns=header) 
df
```

```python
rows=table.find_all("tr") 
for i in rows[1:]: 
    first_td=i.find_all("td")[0].find("span",class_="out-padding").text.strip()
    data=i.find_all("td")[1:] 
    row=[td.text for td in data] 
    row.insert(0,first_td) 
    df.loc[len(df)]=row   
print(df)
```

```python
df.to_csv("OEE2.csv") 
```
